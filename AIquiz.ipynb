```python
import openai
import fitz  # PyMuPDF for PDF text extraction
from keybert import KeyBERT

# Replace with your OpenAI API key
API_KEY = ""

# Configure OpenAI client
client = openai.OpenAI(api_key=API_KEY)

# Function to extract text from a PDF using PyMuPDF
def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text("text") + "\n"
    return text

# Function to extract keywords using KeyBERT
def extract_keywords(text, num_keywords=10):
    kw_model = KeyBERT()
    keywords = kw_model.extract_keywords(text, top_n=num_keywords)
    return [keyword[0] for keyword in keywords]  # Extract only the keyword text

# Function to generate MCQs using OpenAI GPT-4
def generate_mcqs(keywords):
    prompt = f"""
    Generate 10 multiple-choice questions (MCQs) based on the following keywords: {', '.join(keywords)}.
    These MCQs should assess a student's interest in topics related to the keywords.
    Each question should have four answer choices, with one correct answer and three distractors.
    Provide the correct answer for each question in the following format:

    1. Question?
       a) Option 1
       b) Option 2
       c) Option 3
       d) Option 4
       Correct Answer: [Correct Option]
    """

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are an expert quiz generator."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=1000
    )

    return response.choices[0].message.content

pdf_path = "/content/Jagdish_resume.pdf"

resume_text = extract_text_from_pdf(pdf_path)

keywords = extract_keywords(resume_text)

print("\nExtracted Keywords:", keywords)

mcqs = generate_mcqs(keywords)

print("\nGenerated MCQs:\n", mcqs)
```

```python
# Function to conduct the quiz
def conduct_quiz(mcqs):
    score = 0
    total_questions = 0
    topic_scores = {}

    # Split MCQs into question-answer format
    questions = mcqs.strip().split("\n\n")

    for q in questions:
        lines = q.split("\n")
        if len(lines) < 6:
            continue  # Skip malformed questions

        # Extract question and options
        question_text = lines[0]
        options = lines[1:5]
        correct_answer = lines[5].replace("Correct Answer: ", "").strip()

        # Display question
        print(f"\n{question_text}")
        for option in options:
            print(option)

        # Get user's answer
        user_answer = input("Your answer (a, b, c, d): ").strip().lower()

        # Check correctness
        if user_answer == correct_answer[0].lower():  # Comparing first letter
            print("✅ Correct!")
            score += 1
        else:
            print(f"❌ Incorrect! The correct answer was: {correct_answer}")

        total_questions += 1

        # Track topic-based scores (assuming the keyword is in the question)
        for keyword in keywords:
            if keyword.lower() in question_text.lower():
                topic_scores[keyword] = topic_scores.get(keyword, 0) + (1 if user_answer == correct_answer[0].lower() else 0)

    return score, total_questions, topic_scores
```

```python
# Function to analyze interests from quiz performance
def analyze_interests(topic_scores):
    # Sort topics based on correct answers
    sorted_topics = sorted(topic_scores.items(), key=lambda x: x[1], reverse=True)

    if not sorted_topics:
        return "Not enough data to analyze interests."

    top_topics = [topic for topic, score in sorted_topics[:3]]  # Top 3 topics

    return f"\nYour top areas of interest based on the quiz: {', '.join(top_topics)}"
```

```python
# Function to suggest careers based on performance
def suggest_careers(topic_scores):
    top_interests = sorted(topic_scores.items(), key=lambda x: x[1], reverse=True)[:3]  # Get top 3 interests
    interest_keywords = [t[0] for t in top_interests]

    if not interest_keywords:
        return "No strong interests detected. Try again with another quiz."

    prompt = f"""
    Based on a student's strong performance in topics related to {', '.join(interest_keywords)}, suggest some suitable career paths.
    Provide career suggestions in a concise and professional manner.
    """

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are a career advisor."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500
    )

    return response.choices[0].message.content
```

```python
# 🔹 Conduct the quiz
score, total_questions, topic_scores = conduct_quiz(mcqs)

# 🔹 Show quiz results
print(f"\nYour Score: {score}/{total_questions}")

# 🔹 Analyze the user's interests
interest_analysis = analyze_interests(topic_scores)
print(interest_analysis)

# 🔹 Suggest career paths based on quiz performance
career_suggestions = suggest_careers(topic_scores)
print("\nSuggested Career Paths:\n", career_suggestions)
```
